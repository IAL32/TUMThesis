```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

data = pd.read_csv(os.path.abspath('') + '/gpu-vram-over-time.csv')

sns.set_style("ticks", {'axes.grid': True})
ax = sns.scatterplot(data, x='release_year', y='vram')

ax.set(xlabel='Release Date', ylabel='VRAM (GB)')
ax.set_ylim(0, 100)
ax.set_xticks(range(2016, 2026))
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

ax.yaxis.set_major_locator(ticker.MultipleLocator(8))

for i, point in data.iterrows():
    if point["gpu"] in ["GeForce RTX 3090", "A40"]:
        ax.text(point['release_year'] + 0.1, point['vram'] - 4, point['gpu'])
    else:
        ax.text(point['release_year'] + 0.1, point['vram'] + 1, point['gpu'])
        

ax.figure.savefig('../../figures/gpu-vram-over-time.pdf', bbox_inches="tight")
```
