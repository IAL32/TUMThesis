```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

data = pd.read_csv(os.path.abspath('') + '/gpu-vram-over-time.csv')

fig = plt.figure()

sns.set(rc={'figure.figsize':(12, 8)})
ax = sns.scatterplot(data, x='release_year', y='vram')

ax.set(title='GPU VRAM increase over time', xlabel='Release Date', ylabel='VRAM (GB)')
ax.set_ylim(0, 100)
ax.set_xticks(range(2016, 2026))

ax.yaxis.set_major_locator(ticker.MultipleLocator(8))

for i, point in data.iterrows():
    ax.text(point['release_year'] + 0.1, point['vram'], point['gpu'])

plt.show()

fig.savefig('../../figures/gpu-vram-over-time.pdf')
```
